```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Ma'lumotlarni yuklash
train_data = pd.read_csv('/content/train.csv')
test_data = pd.read_csv('/content/test.csv')
test_target = pd.read_csv('/content/sample_submission.csv')
```

```python
encoder=LabelEncoder()
train_data['Geography']=encoder.fit_transform(train_data['Geography'].values)
train_data['Gender']=encoder.fit_transform(train_data['Gender'].values)
train_data['Surname']=encoder.fit_transform(train_data['Surname'].values)
test_data['Geography']=encoder.fit_transform(test_data['Geography'].values)
test_data['Gender']=encoder.fit_transform(test_data['Gender'].values)
test_data['Surname']=encoder.fit_transform(test_data['Surname'].values)
```

```python
features = ['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance',
            'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']
X = train_data[features]
y = train_data['Exited']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
test_data_scaled = scaler.transform(test_data[features])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Model yaratish va o'qitish
model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_lr.fit(X_train, y_train)

# Bashorat qilish
y_pred_lr = model_lr.predict(X_test)

# Aniqlikni baholash
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Logistic Regression Model Accuracy: {accuracy_lr * 100:.2f}%")
```

```python
from sklearn.tree import DecisionTreeClassifier

# Model yaratish va o'qitish
model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(X_train, y_train)

# Bashorat qilish
y_pred_dt = model_dt.predict(X_test)

# Aniqlikni baholash
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Decision Tree Model Accuracy: {accuracy_dt * 100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Model yaratish va o'qitish
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)

# Bashorat qilish
y_pred_rf = model_rf.predict(X_test)

# Aniqlikni baholash
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Model Accuracy: {accuracy_rf * 100:.2f}%")
```

```python
from sklearn.svm import SVC

# Model yaratish va o'qitish
model_svm = SVC(kernel='linear', random_state=42)
model_svm.fit(X_train, y_train)

# Bashorat qilish
y_pred_svm = model_svm.predict(X_test)

# Aniqlikni baholash
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Model Accuracy: {accuracy_svm * 100:.2f}%")
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Model yaratish va o'qitish
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, y_train)

# Bashorat qilish
y_pred_knn = model_knn.predict(X_test)

# Aniqlikni baholash
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"KNN Model Accuracy: {accuracy_knn * 100:.2f}%")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Model yaratish va o'qitish
model_gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
model_gb.fit(X_train, y_train)

# Bashorat qilish
y_pred_gb = model_gb.predict(X_test)

# Aniqlikni baholash
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print(f"Gradient Boosting Model Accuracy: {accuracy_gb * 100:.2f}%")
```
